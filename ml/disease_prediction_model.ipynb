```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,  precision_score, recall_score, f1_score
from sklearn.model_selection import  RandomizedSearchCV, train_test_split , GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
train_df = pd.read_csv('data/Training.csv')
```

```python
train_df.head()
```

```python
train_df.columns
```

```python
train_df = train_df.drop(['Unnamed: 133'],axis=1)
```

```python
train_df.shape
```

```python
# 4920 rows ans 133 columns
train_df.isnull().sum()
#no missing values
```

```python
#train_df.describe()

#train data assign
x_train = train_df.drop(["prognosis"],axis=1)
y_train = train_df["prognosis"]
```

```python
test_df = pd.read_csv('data/Testing.csv')
test_df.head()
```

```python
x_test = test_df.drop(["prognosis"], axis = 1)  
y_test = test_df["prognosis"] 
```

```python
# Define pipeline
pipe = Pipeline([
    ('columntransformer', ColumnTransformer([
        ('scale', MinMaxScaler(), slice(None))
    ])),
    ('kneighborsclassifier', KNeighborsClassifier())
])
```

```python
parametrs_knn = {
    'kneighborsclassifier__n_neighbors': [3, 5, 7],
    'kneighborsclassifier__weights': ['uniform', 'distance'],
}
```

```python

# Perform GridSearchCV using the pipeline
grid = GridSearchCV(pipe, parametrs_knn, cv=5, scoring='accuracy')
grid.fit(x_train, y_train)
```

```python
# Display the best score and parameters
print("Best GridSearchCV score: ", grid.best_score_)
print("Best GridSearchCV parameters: ", grid.best_params_)
```

```python
y_pred = grid.predict(x_test)
```

```python
ac_knn = accuracy_score(y_test, y_pred)
print("Accuracy score: ", ac_knn)
cr_knn = classification_report(y_test, y_pred)
print("Classification report:\n", cr_knn)
```

```python
# Cross-validation using the pipeline

import pickle
cv_score = cross_val_score(grid.best_estimator_, x_train, y_train, cv=5, scoring='accuracy').mean()
print("Cross-validation score: ", cv_score)
```

```python
# Export the best model pipeline
with open('pipe_knn.pkl', 'wb') as file:
    pickle.dump(grid.best_estimator_, file)
```
